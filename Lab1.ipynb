```python
import numpy as np
from matplotlib import pyplot as plt
def image_augmentation(img, mypath):
    
    
    #trasnformation 1
    new_image = cv2.flip(img, 0)
    cv2.imwrite(mypath+'\img1.jpg',new_image)
    
    #trasnformation 2
    cropped_image = img[80:280,150:330]
    cv2.imwrite(mypath+'\img2.jpg',cropped_image)
    
    #trasnformation 3
    image_rotation = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
    cv2.imwrite(mypath+'\img3.jpg',image_rotation)
  
    #trasnformation 4
    matrix = np.ones(img.shape, dtype = "uint8") * 120 
    image_britness = cv2.add(img, matrix)
    cv2.imwrite(mypath+'\img4.jpg',image_britness)
    
    #tranformation 5
    processed_img = cv2.medianBlur(img,5)
    cv2.imwrite(mypath+'\img5.jpg',processed_img)
    
    #transformation 6
    Kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
    sharpened = cv2.filter2D(img, -1, 5)
    cv2.imwrite(mypath+'\img6.jpg',sharpened)
    
    plt.subplot(2, 4,1), plt.imshow(img),plt.title("original")
    plt.subplot(2, 4,2), plt.imshow(new_image),plt.title("flipped")
    plt.subplot(2, 4,3), plt.imshow(cropped_image),plt.title("cropped_image") 
    plt.subplot(2, 4,4), plt.imshow(image_rotation),plt.title("image_rotation")
    plt.subplot(2, 4,5), plt.imshow(image_britness),plt.title("image_britness")
    plt.subplot(2, 4,6), plt.imshow(processed_img),plt.title("processed_img")
    plt.subplot(2, 4,7), plt.imshow(sharpened),plt.title("sharpened")
    plt.show()

   
  
```

```python
image_augmentation(image, path_save)
```
